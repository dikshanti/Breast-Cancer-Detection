# Violin Shape:

The shape of the violin represents the distribution of the radius_mean for each diagnosis category.
The width of the violin at any given point represents the density of the data at that value. Wider sections indicate where more data points are concentrated.

# Black Box Plot Inside the Violin:

Each violin contains a black box plot showing key statistical values:
The white dot is the median value.
The thick bar represents the interquartile range (IQR), i.e., the middle 50% of the data.
The thin line extending from the bar (called whiskers) shows the overall spread of the data (excluding outliers).

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
breast_cancer['diagnosis'] = breast_cancer['diagnosis'].map({'M': 1, 'B': 0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.drop('diagnosis',axis=1), 
                                                    breast_cancer['diagnosis'], test_size=0.30, 
                                                    random_state=101)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
prediction = logmodel.predict(X_test)
```

```python
predictions_df = pd.DataFrame(prediction, columns=['prediction'])
```

```python
# Create a DataFrame for the actual values
y_test_df = pd.DataFrame(y_test.values, columns=['actual'])
```

```python
combined_df = pd.concat([predictions_df, y_test_df.reset_index(drop=True)], axis=1)
```

```python
sns.countplot(x='prediction',data=combined_df,alpha=0.8)
sns.countplot(x='actual',data=combined_df,alpha=0.6)
```

```python
from sklearn.metrics import classification_report
```

```python
x = classification_report(y_test, prediction)
print(x)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,prediction))
```

## K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,pred))
```

```python
y = classification_report(y_test,pred)
print(y)
```

```python
from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_knn = accuracy_score(y_test, pred)
print(f'Accuracy: {accuracy_knn * 100:.2f}%')
```

## Support Vector Machine (SVM)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the model
svm_classifier.fit(X_train, y_train)

# Make predictions
y_pred = svm_classifier.predict(X_test)
```

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_svm = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy_svm * 100:.2f}%')
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree_classifier = DecisionTreeClassifier(random_state=42)

# Train the model
decision_tree_classifier.fit(X_train, y_train)

# Make predictions
y_prediction = decision_tree_classifier.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_prediction))

print("\nClassification Report:")
print(classification_report(y_test, y_prediction))
```

```python
from sklearn.metrics import accuracy_score

accuracy_dt = accuracy_score(y_test, y_prediction)
print(f'Accuracy: {accuracy_dt * 100:.2f}%')
```

## Accuracy of each model

```python
from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_log_reg = accuracy_score(y_test, prediction)
print(f'Logistic Regression: {accuracy_log_reg * 100:.2f}%')

from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_knn = accuracy_score(y_test, pred)
print(f'K Nearest Neighbours: {accuracy_knn * 100:.2f}%')

from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_svm = accuracy_score(y_test, y_pred)
print(f'Support Vector Machine: {accuracy_svm * 100:.2f}%')

from sklearn.metrics import accuracy_score
# Accuracy score
accuracy_dt = accuracy_score(y_test, y_prediction)
print(f'Decision Trees: {accuracy_dt * 100:.2f}%')
```

```python
# Accuracy values for each model
model_names = ['Logistic Regression', 'K Nearest Neighbours', 'SVM', 'Decision Trees']
accuracies = [accuracy_log_reg, accuracy_knn, accuracy_svm, accuracy_dt]

# Create a DataFrame for better handling with Seaborn
accuracy_df = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracies
})

accuracy_df['Accuracy'] = accuracy_df['Accuracy'] * 100

plt.figure(figsize=(10, 8))
bar_plot = sns.barplot(x='Model', y='Accuracy', data=accuracy_df, palette='Blues_d')

# Annotating the bars with accuracy percentages
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.2f}%',
                      (p.get_x() + p.get_width() / 2., p.get_height()-1), 
                      ha='center', va='center', 
                      xytext=(0, 9),  # Offset for better visibility
                      textcoords='offset points', fontsize=11)
    
# Set y-axis limits from 0 to 100
plt.ylim(0, 100)

# Add title and labels
plt.title('Model Accuracy Comparison', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.xlabel('Model', fontsize=12)
plt.xticks(rotation=15)
plt.show()
```
